# I rounded to nearest integer, as Illumin does the same with quality scores. This saves memory, as a float is a greater sink than an int

## Quality cutoff score:
At a quality score of 30, there is a 1 in 1000 chance that a base is incorrect. This holds true for each of the scenarios, as the likelifrom matplotlib import pyplot as plt
As indices are the identifying factor for the sequences, the quality must be higher, so >= 33 would be a better cutoff, since it is the more important part

```python
from matplotlib import pyplot as plt
x = []
with open("../seq1distscores.out", 'r') as seq:
    for line in seq:
        x.append(float(line))
plt.plot(x)
plt.xlabel("position")
plt.ylabel("score")
plt.show
```
